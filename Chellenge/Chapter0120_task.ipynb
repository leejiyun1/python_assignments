```python
# 문제 풀이
# 문제 1
class BungeoppangShop: 
    def __init__(self):
        self.stock = {
            "팥붕어빵" : 10,
            "슈크림붕어빵" : 8,
            "초코붕어빵" : 5
        }
        self.prices = {
            "팥붕어빵" : 1000,
            "슈크림붕어빵" : 1200,
            "초코붕어빵" : 1500
        }
        self.sales = {
            "팥붕어빵" : 0,
            "슈크림붕어빵" : 0,
            "초코붕어빵" : 0
        }
        
    # 문제 2
    def check_stock(self):
        for bread_type, bread_count in self.stock.items():
            print(f"{bread_type}: {bread_count}개")

        # 문제 3
    def process_order(self, bread_type, bread_count):
            if self.stock[bread_type] >= bread_count : # 재고 확인
                self.stock[bread_type] -= bread_count # 차감
                self.sales[bread_type] += bread_count # 주문량 증가
                print(f"{bread_type}:{bread_count}개 입니다. 남은 재고는 {self.stock[bread_type]}개 입니다.")
            else:
                print(f"{bread_type}의 재고가 부족합니다.")
            
    # 문제 4
    def admin_mode(self):
        print("관리자 모드 입니다.")
        print("1. 현재 재고 확인")
        print("2. 재고 추가")
        print("3. 종료")

        while True:
            admin_number = input("메뉴를 선택하세요: ")
            if admin_number == "1":
                self.check_stock()

            elif admin_number == "2":
                while True:
                    bread_type = input("어떤 붕어빵을 추가하겠습니까? (종료하려면 '종료' 입력)")

                    if bread_type in self.stock:
                        additional_stock = int(input("개수를 입력해주세요."))
                        self.stock[bread_type] += additional_stock
                        print(f"{bread_type}:{additional_stock}개를 추가 하셨습니다. 총재고는 {self.stock[bread_type]}개 입니다.")

                    elif bread_type == "종료":
                        print("관리자 모드를 종료합니다.")
                        return

                    else:
                        print("잘못된 입력입니다.")


            elif admin_number == "3":
                print("관리자 모드를 종료합니다.")
                break

            else:
                print("잘못된 입력입니다.")
    # 문제 5번	    
    def calculate_total_sales(self):
            total_sales = sum(self.sales[bread_type] * self.prices[bread_count] for bread_type in self.sales)
            print(f"총 매출: {total_sales}원")

#문제 6번
def main():
    shop = BungeoppangShop()

    while True:
        order_type = input("모드를 선택해주세요 1.주문 2.관리자 모드 3. 종료")
        if order_type == "1":
                bread_type = input("붕어빵 주문 도와드리겠습니다. 종류를 입력해주세요. (종료하려면 '종료' 입력) ")
            
                if bread_type in shop.stock:
                    bread_count = int(input("개수를 입력해주세요: "))
                    shop.process_order(bread_type, bread_count)  
            
                elif bread_type == "종료":  
                    print("주문을 종료합니다.")
                    break
                else:
                    print("없는 붕어빵 입니다.(팥붕어빵, 슈크림붕어빵, 초코붕어빵)중에 입력해주세요.")

        elif order_type == "2":
            admin_pw = input("암호를 입력해주세요.")

            if admin_pw == "admin123":
                 shop.admin_mode()
            else:
                print("암호가 다릅니다. 종료합니다.")
                break
            
        elif order_type == "3":
            print("종료합니다.")
            break

        else:
            print("다시 입력해주세요.")

if __name__ == "__main__":
    main()

```
